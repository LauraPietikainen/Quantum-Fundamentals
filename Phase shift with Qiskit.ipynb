```python
from qiskit import *
from qiskit.tools.visualization import plot_bloch_multivector
%matplotlib inline
```

```python
circuit = QuantumCircuit(3)
circuit.h([0, 1, 2]) #3 Qubits with Hadamard gate
circuit.t(1) #T-gate to middle qubit rot. 45 dergee
circuit.tdg(2) #T-dagger gate rot. neg 45 degree
# circuit.s(1) #S-gate to middle qubit rot. 90 degree
# circuit.sdg(2) #S-dagger gate rot. neg 90 degree
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('statevector_simulator')
result = execute(circuit, backend=simulator).result()
plot_bloch_multivector(result.get_statevector())
```
