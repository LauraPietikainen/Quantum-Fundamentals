```python
from qiskit import *
from qiskit.tools.visualization import plot_bloch_multivector
from math import pi
%matplotlib inline
```

```python
circuit = QuantumCircuit(3)
circuit.x(0) #Toffoli doesn't flip the target unless both control qubits are one
circuit.x(1) #2nd X gate to demonstrate same control state, target flips
circuit.toffoli(0,1,2) #CCNOT 1st and 2nd Control, 3rd Target
#circuit.ccx(0,1,2) #CCX = Controlled controlled NOT, same as Toffoli
circuit.rx(pi/3,2) #To demonstrate that Toffoli gate rotated the target 180 degrees on X axis
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('statevector_simulator')
result = execute(circuit, backend=simulator).result()
plot_bloch_multivector(result.get_statevector())
```
