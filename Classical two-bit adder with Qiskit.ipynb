```python
from qiskit import *
from qiskit.tools.visualization import plot_bloch_multivector
from math import pi
%matplotlib inline
```

```python
circuit = QuantumCircuit(4, 2)
circuit.x(0) #initialize input A
circuit.x(1) #initialize input B
circuit.barrier()

circuit.cx(0, 2) #CNOT to flip the sum
circuit.cx(1, 2) #CNOT, uses second input qubit to flip the sum
circuit.ccx(0, 1, 3)

circuit.barrier()
circuit.measure(2, 0) #measure SUM
circuit.measure(3, 1) #measure carry-out
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('statevector_simulator')
result = execute(circuit, backend=simulator).result()
plot_bloch_multivector(result.get_statevector())
```
