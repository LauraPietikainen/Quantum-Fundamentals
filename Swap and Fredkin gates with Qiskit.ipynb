```python
from qiskit import *
from qiskit.tools.visualization import plot_bloch_multivector
from math import pi
%matplotlib inline
```

```python
circuit = QuantumCircuit(3)
circuit.x(0) #Control gate to one, Fredkin gate swaps the other two qubits
circuit.h(1) #Hadamard gate
circuit.rx(-pi/4, 2)
#circuit.swap(1, 2) #swap gate, swaps the states of the last 2 qubits
#circuit.fredkin(0, 1, 2) #Fredkin Controlled swap (Control, Target, Target)
circuit.cswap(0, 1, 2) #Same as Fredkin
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('statevector_simulator')
result = execute(circuit, backend=simulator).result()
plot_bloch_multivector(result.get_statevector())
```
