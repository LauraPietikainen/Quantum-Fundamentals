```python
from qiskit import *
from qiskit.tools.visualization import plot_bloch_multivector
from math import pi
%matplotlib inline
```

```python
circuit = QuantumCircuit(3)
circuit.rx(2, 0) #rotation 2 radius / 120 degrees
# circuit.rx(pi/3, 0) #first qubit: rotation to index 0 x-axis
circuit.ry(-pi/3, 1) #Middle qubit: negative pi over three rotation around y-axis
circuit.h(2) #Hadamard gate
circuit.rz(pi/3, 2) #3rd qubit, pi over three radians around the z-axis
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('statevector_simulator')
result = execute(circuit, backend=simulator).result()
plot_bloch_multivector(result.get_statevector())
```
