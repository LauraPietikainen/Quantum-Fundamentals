```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram, array_to_latex, plot_bloch_multivector
%matplotlib inline
```

```python
# To register one Qubit
# qr = QuantumRegister(2) #growing circuit from 1->2 bits
# To store one classical bit
# cr = ClassicalRegister(2) #similar to cirquit bits
# Combine those two registers into a quantum circuit, passing into quantum register
# circuit = QuantumCircuit (qr, cr)
circuit = QuantumCircuit(2)
# Hadamard gate (50% change for 0 of being measured as 1 or 0)
# circuit.h(qr) # horisontal vector pointing to X-axes
circuit.h(0)
# Using that cirquit will perform a measure operation on the quantum register and store the result.. 
# ..Of measuring the quantum register in the classical register
# circuit.measure(qr, cr) #Points to result 0 or 1
circuit.measure_all()
```

```python
# Create a new simulator (QASM simulation)
# QASM = Quantum Assembly Language
# simulator = Aer.get_backend('statevector_simulator')
simulator = Aer.get_backend('qasm_simulator')
# Setting the backend for simulator, input argument 1024 shots (how many times to run the program)
# Outcome of measuring a qubit in superpositio is not deterministic
# It's random according to some distribution of probabilities
result = execute(circuit, backend = simulator).result() #state_simulator
# result = execute(circuit, backend=simulator, shots=1024).result()
# statevector = result.get_statevector()
counts = result.get_counts()
# print(statevector)
# array_to_latex(statevector)
# plot_histogram(result.get_counts())
```
